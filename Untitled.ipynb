# Primerjava ocen filmov

Ideja projekta je, da raziščemo povezavo med ocenami filmov, igralcev, žanrov in let in s tem poskusimo dobiti oziroma si ustvariti neko idejo za dobro formulo za preračunavanje najljubših igralcev, režiserjev in žanrov.

```python
import pandas as pd
```

Najprej preberimo uporabnikovo bazo filmov in si poglejmo, kako ta sploh izgleda.

```python
filmbase = pd.read_csv('ur33974390.csv', sep='|')
filmbase
```

## Analiza podatkov

Predno se lotimo iskanja formule si raje poglejmo njegove najljubše filme, katera obdobja so uporabniku bolj všeč, njegove omiljene igralke in igralce ter žanre. S pomočjo tega bomo na koncu lažje preverili učinkovitost priporočevalca, poleg tega pa bomo podobne ukaze rabili potem, ko bomo želeli pridobiti podatke o najljubših filmih in igralcih.

```python
filmbase[['title', 'year', 'grade']].sort_values(['grade'], ascending=False).head(30)
```

*Zelo sem presenečen, da imam več kot 30 filmov ocenjenih z 9... - op.a.*

In še najslabših 30...

```python
filmbase[['title', 'year', 'grade']].sort_values(['grade'], ascending=True).head(30)
```

Kakšne ocene pa kaj daje naš uporabnik?

```python
%matplotlib inline

filmi_po_ocenah = filmbase.groupby('grade')
filmi_po_ocenah.size().plot(kind='pie', title='Grades distribution')
```

Poglejmo si, koliko filmov iz določenega desetletja je uporabnik pogledal. Nato izračunajmo še povprečno oceno desetletja.

```python
filmbase['desetletje'] = 10 * (filmbase.year // 10)
filmi_po_desetletjih = filmbase.groupby('desetletje')
filmi_po_desetletjih.size()
```

```python
ocene_po_desetletjih = filmi_po_desetletjih.mean()['grade'].round(2)
ocene_po_desetletjih
```

```python
%matplotlib inline
ocene_po_desetletjih.plot(kind='bar')
```

Poiščimo povprečne ocene filmov za posamezne žanre, igralce in režiserje.

```python
genres = pd.read_csv('genres.csv', sep='|')
genres.join(filmbase['grade'], on='imdb_id')[['genre', 'grade']]\
    .groupby('genre').mean().round(2).sort_values(['grade'], ascending=False)
```

```python
actors = pd.read_csv('actors.csv', sep="|")
igralci_po_ocenah = actors.join(filmbase['grade'], on='imdb_id')[['actor', 'grade']]\
    .groupby('actor').mean().round(2).sort_values(['grade'], ascending=False)
igralci_po_ocenah
```

```python
directors = pd.read_csv('directors.csv', sep="|")
reziserji_po_ocenah = directors.join(filmbase['grade'], on='imdb_id')[['director', 'grade']]\
    .groupby('director').mean().round(2).sort_values(['grade'], ascending=False)
reziserji_po_ocenah
```

Seveda je pričakovati, da rezultati ne bodo ravno objektivni, saj je lahko nek igralec igral samo v enem filmu, ki si ga je uporabnik ogledal in ga je ocenil z 10, kar pa ga seveda ne naredi najbolj priljubljenega igralca.

Pojdimo se igrat "statistiko". Poglejmo, kakšno je razmerje med ocenami igralcev in desetletij, v katerih so igrali.

```python
igralci_po_desetletjih = actors.join(filmbase['desetletje'], on='imdb_id')[['actor', 'desetletje']]\
    .groupby('actor').mean().round(2)
igralci_po_desetletjih
```

```python
igralci_zdruzeno = igralci_po_desetletjih.join(igralci_po_ocenah)\
    .join(filmi_po_desetletjih.mean()['grade'].round(2), on='desetletje', rsuffix='_desetletja').sort_values('grade', ascending=False)
igralci_zdruzeno
```

```python
reziserji_po_desetletjih = directors.join(filmbase['desetletje'], on='imdb_id')[['director', 'desetletje']]\
    .groupby('director').mean().round(2)
reziserji_po_desetletjih
```

```python
reziserji_zdruzeno = reziserji_po_desetletjih.join(reziserji_po_ocenah)\
    .join(filmi_po_desetletjih.mean()['grade'].round(2), on='desetletje', rsuffix='_desetletja').sort_values('grade', ascending=False)
reziserji_zdruzeno
```

Pričakovano ocene tu zelo odstopajo, saj je lahko bil v enem obdobju določen film odličen, ostali pa so bili bolj zanič. Prav ta odličen film pa je dvigoval oceno predvsem igralcem, ki si nastopali samo v tem filmu.

Za zaključek si poskusimo na podlagi ugotovitev zgoraj izmisliti čim primernejšo formulo. Glavno vlogo v njej morajo seveda igrati ocene filmov, vendar bi bilo po mojem mnenju potrebno dati večjo težo številu ocen določenega igralca, režiserja oziroma žanra, kot pa dejanski oceni, saj je moje mnenje tako, da če igralec, režiser oziroma žanr gledalcu ni všeč, bo pogledal manj takih filmov. Poleg tega je bila nekomu všeč srhljiva komedija, pa zaradi tega še ne mara shrljivk, saj je bil tisti film bolj komičen kot srhljiv in mu je bil zato všeč. Vendar v nasprotju temu lahko rečemo, da se ponekod pojavlja en in isti igralec v filmih, ki so gledalcu všeč, vendar so mu všeč zaradi česa drugega in ne zaradi njega. Zato je to formulo težko najti, če pa bi se moral za kaj odločiti, bi pa gledalce in režiserje najverjetneje obračunal po naslednji formuli

*vsota(ocena_filma* x *povprečje_ocen_filmu_pripadajočih_žanrov * število_filmov_iz_desetleja) / število_vseh_filmov_igralca + število_vseh_filmov_igralca / število_vseh_filmov*.
